```python
from langchain_core.tools import tool

@tool
def division(a: int, b: int) -> int:
    """Returns the result of dividing a by b"""
    return a/b
```

```python
print(division.name)
print(division.description)
print(division.args)
```

```python
@tool
async def multiple (a: int, b: int) -> int:
    """Returns the result of multiply a by b"""
    return a*b
```

```python
print(multiple.name)
print(multiple.description)
print(multiple.args)
```

```python
from typing import Annotated, List

@tool
async def multiply_by_max(
    a: Annotated[int, "The first number to multiply"],
    b: Annotated[List[int],"list of values"]
) -> int:
    """Returns the result of multiply a by b"""
    return a*max(b)
```

```python
print(multiply_by_max.name)
print(multiply_by_max.description)
print(multiply_by_max.args)
```

```python
### Structured tool
from langchain_core.tools import StructuredTool
def multiple (a: int, b: int) -> int:
    """Returns the result of multiply a by b"""
    return a*b
async def amultiple (a: int, b: int) -> int:
    """Returns the async result of multiply a by b"""
    return a*b
calcutaor= StructuredTool.from_function(func=multiple,coroutine=amultiple)



print(calcutaor.name)
print(calcutaor.description)
print(calcutaor.args)

print(calcutaor.invoke({"a":2,"b":3}))
print(await calcutaor.arun({"a":2,"b":5}))
```

```python
from google.colab import userdata
groq_api_key=userdata.get('groq_api_key')
groq_api_key[:10]
```

```python
from langchain.chat_models import init_chat_model
llm=init_chat_model("qwen/qwen3-32b",model_provider="groq", groq_api_key=groq_api_key)
llm
```

```python
from langchain_community.tools import WikipediaQueryRun, ArxivQueryRun
from langchain_community.utilities import WikipediaAPIWrapper, ArxivAPIWrapper

# Re-initialize api_wrapper and wiki_tool to ensure they are available within this cell's context
api_wrapper = WikipediaAPIWrapper(top_k_results=5, doc_content_chars_max=5000)
wiki_tool = WikipediaQueryRun(api_wrapper=api_wrapper)

# Correctly initialize arxiv_tool
arxiv_tool = ArxivQueryRun(api_wrapper=ArxivAPIWrapper())

# Re-create the tools list with the corrected arxiv_tool
tools = [wiki_tool,calcutaor]
for t in tools:
  print(t.name)

llm_with_tools=llm.bind_tools(tools)

llm_with_tools



```

```python
query="what is 2 * 3"

response= llm_with_tools.invoke(query)
response.tool_calls
```

```python
message=[]
for tool_call in response.tool_calls:
  selected_tools={"multiple":calcutaor,"wikipedia":wiki_tool}[tool_call["name"].lower()]
  tool_msg= selected_tools.invoke(tool_call)
  message.append(tool_msg)

  # print(tool_call)
print(message)
```

```python
from langchain_core.messages import HumanMessage
query="what is 2 * 3"
messages = [HumanMessage(query)]

response= llm_with_tools.invoke(query)

# response.tool_calls
for tool_call in response.tool_calls:
  selected_tools={"multiple":calcutaor,"wikipedia":wiki_tool}[tool_call["name"].lower()]
  tool_msg= selected_tools.invoke(tool_call)
  messages.append(tool_msg)

res=llm_with_tools.invoke(messages)
res
res.content
```
